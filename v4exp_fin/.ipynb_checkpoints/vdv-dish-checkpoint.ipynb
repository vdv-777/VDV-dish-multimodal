Разведывательный анализ датасета для обучения

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load
ingr = pd.read_csv("data/ingredients.csv")
dish = pd.read_csv("data/dish.csv")

# 1. Calories distribution
plt.figure(figsize=(10,4))
sns.histplot(dish["total_calories"], kde=True)
plt.title("Distribution of total_calories")
plt.show()

# 2. Number of ingredients per dish
dish["n_ingredients"] = dish["ingredients"].apply(lambda x: len(x.split(";")) if pd.notna(x) else 0)
sns.histplot(dish["n_ingredients"])
plt.title("Number of ingredients per dish")
plt.show()

# 3. Top-20 ingredients
all_ingr = []
for row in dish["ingredients"].dropna():
    ids = row.split(";")
    nums = [int(i.replace("ingr_", "")) for i in ids if i.startswith("ingr_")]
    all_ingr.extend(nums)
top_ids = pd.Series(all_ingr).value_counts().head(20)
top_names = [ingr[ingr["id"] == id]["ingr"].iloc[0] for id in top_ids.index]
print("Top ingredients:", top_names)
```

Инференс

```python
from config import Config
from model import CaloriePredictionModel
from inference import run_inference
from train import prepare_ingredient_mapping  # ← используем ту же функцию
import pandas as pd
import torch

# Настройка
cfg = Config()
device = "cuda" if torch.cuda.is_available() else "cpu"

# === Подготовка ingr_to_idx (как в train.py) ===
ingredients_df = pd.read_csv("data/ingredients.csv")
dish_df = pd.read_csv("data/dish.csv")
dish_df = prepare_ingredient_mapping(ingredients_df, dish_df)

all_names = set()
for names in dish_df["ingredient_names"]:
    all_names.update(names)
ingr_to_idx = {name: idx + 1 for idx, name in enumerate(sorted(all_names))}
ingr_to_idx["<PAD>"] = 0

# === Загрузка модели ===
model = CaloriePredictionModel(cfg, num_ingredients=len(ingr_to_idx))
model.load_state_dict(torch.load(cfg.MODEL_SAVE_PATH, map_location="cpu"))
model.to(device).eval()

# === Загрузка инференс-данных ===
inference_df = pd.read_csv("vdv-dish/dish.csv")
inference_df.columns = inference_df.columns.str.strip()

# Обязательно: колонка 'ingredients' и желательно 'mass'
if "ingredients" in inference_df.columns:
    inference_df["ingredient_names"] = inference_df["ingredients"].apply(
        lambda x: x.split(",") if pd.notna(x) else []
    )
else:
    inference_df["ingredient_names"] = [[] for _ in range(len(inference_df))]

# Колонка 'mass' — если не указана, будет использована заглушка (218.0 в dataset.py)

# === Запуск инференса ===
results = run_inference(cfg, model, ingr_to_idx, inference_df, device)

# === Сохранение ===
output_path = "vdv-dish/predictions.csv"
results.to_csv(output_path, index=False)
print(f"✅ Инференс завершён. Результаты сохранены в: {output_path}")
print(f"   Обработано блюд: {len(results)}")
```

Сохранение эмбеддингов датасета (для возможности дальнейшего инференса без папки data/images/ с картинками обучающего датасета)

```python
# save_inference_bundle.py
import os
import torch
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader
from tqdm import tqdm
from config import Config
from model import MultimodalCalorieModel
from dataset import CalorieDataset, get_transforms
from train import prepare_ingredient_mapping

def main():
    cfg = Config()
    device = "cuda" if torch.cuda.is_available() else "cpu"
    
    # Загрузка обученной модели
    print("Загрузка модели...")
    checkpoint = torch.load(cfg.MODEL_SAVE_PATH, map_location="cpu")
    ingr_to_idx = checkpoint["ingr_to_idx"]
    model = MultimodalCalorieModel(cfg, num_ingredients=len(ingr_to_idx))
    model.load_state_dict(checkpoint["model_state_dict"])
    model.to(device).eval()

    # Подготовка обучающих данных
    print("Загрузка обучающего набора...")
    ingredients_df = pd.read_csv(cfg.INGREDIENTS_PATH)
    dish_df = pd.read_csv(cfg.DISH_PATH)
    train_df, _ = prepare_ingredient_mapping(ingredients_df, dish_df)
    train_df = train_df[train_df["split"] == "train"]

    # Вычисление эмбеддингов
    print("Вычисление эмбеддингов...")
    transforms = get_transforms(cfg, is_train=False)
    dataset = CalorieDataset(train_df, ingr_to_idx, cfg.IMAGES_DIR, transforms, cfg.MAX_INGREDIENTS, mode="train")
    loader = DataLoader(dataset, batch_size=64, shuffle=False, num_workers=4)

    image_embeddings = []
    multimodal_embeddings = []
    dish_ids = []
    calories = []
    masses = []
    ingredient_names_list = []

    with torch.no_grad():
        for image, ingr, cal, mass, did in tqdm(loader, desc="Processing train blds"):
            image, ingr = image.to(device), ingr.to(device)
            _, img_emb, text_emb = model(image, ingr)
            
            image_embeddings.append(img_emb.cpu().numpy())
            fused = torch.cat([img_emb, text_emb], dim=1)
            multimodal_embeddings.append(fused.cpu().numpy())
            
            dish_ids.extend(did)
            calories.extend(cal.numpy())
            masses.extend(mass.numpy())
            for i in range(len(did)):
                row = train_df[train_df["dish_id"] == did[i]].iloc[0]
                ingredient_names_list.append(row["ingredient_names"])

    image_embeddings = np.vstack(image_embeddings)
    multimodal_embeddings = np.vstack(multimodal_embeddings)

    # Сохранение
    bundle = {
        "image_embeddings": image_embeddings,
        "multimodal_embeddings": multimodal_embeddings,
        "dish_ids": dish_ids,
        "calories": calories,
        "masses": masses,
        "ingredient_names": ingredient_names_list,
        "ingr_to_idx": ingr_to_idx,
        "config": cfg
    }
    torch.save(bundle, cfg.EMBEDDINGS_PATH)
    print(f"✅ Inference bundle сохранён в: {cfg.EMBEDDINGS_PATH}")

if __name__ == "__main__":
    main()
```

Отчет ТОП худших и лучших блюд на валидации по обученной модели

```python
import os
import torch
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader
from tqdm import tqdm
from config import Config
from model import MultimodalCalorieModel
from dataset import CalorieDataset, get_transforms
from train import prepare_ingredient_mapping

def main():
    cfg = Config()
    device = "cuda" if torch.cuda.is_available() else "cpu"
    
    # === НАСТРОЙКА: сколько позиций показывать в топах ===
    TOP_K = 5  # ← можно изменить

    # Загрузка модели
    print("🔧 Загрузка модели...")
    checkpoint = torch.load(cfg.MODEL_SAVE_PATH, map_location="cpu")
    ingr_to_idx = checkpoint["ingr_to_idx"]
    model = MultimodalCalorieModel(cfg, num_ingredients=len(ingr_to_idx))
    model.load_state_dict(checkpoint["model_state_dict"])
    model.to(device).eval()

    # Загрузка данных
    print("📊 Загрузка валидационного набора...")
    ingredients_df = pd.read_csv(cfg.INGREDIENTS_PATH)
    dish_df = pd.read_csv(cfg.DISH_PATH)
    dish_df, _ = prepare_ingredient_mapping(ingredients_df, dish_df)
    
    val_df = dish_df[dish_df["split"] == "test"]
    print(f"   Валидационных блюд: {len(val_df)}")

    # DataLoader
    transforms = get_transforms(cfg, is_train=False)
    val_dataset = CalorieDataset(val_df, ingr_to_idx, cfg.IMAGES_DIR, transforms, cfg.MAX_INGREDIENTS, mode="train")
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4)

    # Предсказание
    print("📈 Вычисление предсказаний...")
    val_preds, val_trues, val_dish_ids = [], [], []
    
    with torch.no_grad():
        for image, ingr, calories, mass, dish_id in tqdm(val_loader, desc="Inference on val"):
            image, ingr, calories = image.to(device), ingr.to(device), calories.to(device)
            pred, _, _ = model(image, ingr)
            val_preds.extend(pred.cpu().numpy())
            val_trues.extend(calories.cpu().numpy())
            val_dish_ids.extend(dish_id)

    # Формируем список ошибок
    errors = []
    for p, t, did in zip(val_preds, val_trues, val_dish_ids):
        abs_err = abs(p - t)
        errors.append({
            "dish_id": did,
            "true": t,
            "pred": p,
            "abs_error": abs_err
        })

    # Сортировка
    errors_sorted = sorted(errors, key=lambda x: x["abs_error"], reverse=True)
    worst_top = errors_sorted[:TOP_K]
    best_top = sorted(errors, key=lambda x: x["abs_error"])[:TOP_K]

    # Вывод
    report_lines = []
    
    # === ХУДШИЕ ===
    print("\n" + "="*100)
    print(f"🔍 ТОП-{TOP_K} БЛЮД С НАИБОЛЬШЕЙ ОШИБКОЙ ПРЕДСКАЗАНИЯ")
    print("="*100)
    report_lines.append(f"ТОП-{TOP_K} ХУДШИХ ПРЕДСКАЗАНИЙ:\n" + "-" * 60)
    
    for i, item in enumerate(worst_top, 1):
        did = item["dish_id"]
        true_val = item["true"]
        pred_val = item["pred"]
        err = item["abs_error"]
        
        row = dish_df[dish_df["dish_id"] == did]
        if not row.empty:
            ingr_list = row.iloc[0]["ingredient_names"]
            ingr_str = ", ".join(ingr_list) if isinstance(ingr_list, list) else str(ingr_list)
        else:
            ingr_str = "<ингредиенты не найдены>"
        
        print(f"\n{i}. dish_id: {did}")
        print(f"   Истина: {true_val:7.1f} ккал | Предсказание: {pred_val:7.1f} ккал | Ошибка: {err:7.1f} ккал")
        print(f"   Ингредиенты: {ingr_str}")
        
        report_lines.append(f"{i}. dish_id={did} | ошибка={err:.1f} ккал")
        report_lines.append(f"   Ингредиенты: {ingr_str}")
        report_lines.append("")

    # === ЛУЧШИЕ ===
    print("\n" + "="*100)
    print(f"✅ ТОП-{TOP_K} БЛЮД С НАИЛУЧШИМ ПРЕДСКАЗАНИЕМ")
    print("="*100)
    report_lines.append(f"\nТОП-{TOP_K} ЛУЧШИХ ПРЕДСКАЗАНИЙ:\n" + "-" * 60)
    
    for i, item in enumerate(best_top, 1):
        did = item["dish_id"]
        true_val = item["true"]
        pred_val = item["pred"]
        err = item["abs_error"]
        
        row = dish_df[dish_df["dish_id"] == did]
        if not row.empty:
            ingr_list = row.iloc[0]["ingredient_names"]
            ingr_str = ", ".join(ingr_list) if isinstance(ingr_list, list) else str(ingr_list)
        else:
            ingr_str = "<ингредиенты не найдены>"
        
        print(f"\n{i}. dish_id: {did}")
        print(f"   Истина: {true_val:7.1f} ккал | Предсказание: {pred_val:7.1f} ккал | Ошибка: {err:7.1f} ккал")
        print(f"   Ингредиенты: {ingr_str}")
        
        report_lines.append(f"{i}. dish_id={did} | ошибка={err:.1f} ккал")
        report_lines.append(f"   Ингредиенты: {ingr_str}")
        report_lines.append("")

    # Сохранение в файл
    report_path = os.path.join(cfg.SAVE_DIR, "error_analysis.txt")
    with open(report_path, "w", encoding="utf-8") as f:
        f.write("Анализ точности модели на валидационном наборе\n")
        f.write("="*60 + "\n")
        f.write("\n".join(report_lines))
    print(f"\n📄 Полный отчёт сохранён: {report_path}")

if __name__ == "__main__":
    main()
```
